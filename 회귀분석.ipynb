## 회귀분석의 기본가정
- 독립변수 x는 비확률적이다
 - 오차항 e는 평균 0을 가지는 확률변수이다
 - 오차항 e는 모든 x에 대하여 동일한 분산을 지닌다(동분산/균분산)
 - 오차항 e는 서로 상관되어 있지 않다(비자기상관)
 - 독립변수들간에 완전한 선형관계가 성립되지 않는다(비다중공선성)
 - 오차항 e는 정규분포를 따라야 한다
 
## 최소제곱 추정량
  - 잔차제곱들의 합을 최소화 하는 직선을 선택하는 방법
      - 가우스 마코프의 정리 : 회귀모형의 기본가정들이 만족되는 경우 최소제곱 추정량은 선형불편 추정량들 중 가장 작은 분산을 가진다
  
## 잔차의 정규성 탐지
 - 잔차의 히스토그램이 좌우 대칭 & 종모양 & 하나의 봉우리
 - Jargue Bera 검정통계량
     - 검정통계량 LM이 카이제곱 분포의 임계값을 초과하면 오차항이 정규분포 한다는 귀무가설 기각 (=오차항이 비정규분포한다)
